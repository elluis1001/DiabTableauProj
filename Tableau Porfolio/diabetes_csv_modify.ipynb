```python
# Import Dependencies:
import pandas as pd
```

```python
#  Path to csv for creating it to a working DataFrame:
df = pd.read_csv(r"C:\Users\ellui\OneDrive\Documents\Data bootcamp 2\Tableau Porfolio\Resources\diabetes_prediction_dataset.csv")
df.head()
```

```python
# Get the unique values in a specific column
column_name = 'hypertension'
unique_values = df[column_name].unique()

# Print the unique values
print(unique_values)
```

```python
# Assign strings to the the hypertension values:
# Mapping for replacement
replacement_map = {0: 'not hypertensive', 1: 'hypertensive'}

# Replace values in the column
df['hypertension'] = df['hypertension'].replace(replacement_map)

# Print the updated DataFrame
df.head()
```
